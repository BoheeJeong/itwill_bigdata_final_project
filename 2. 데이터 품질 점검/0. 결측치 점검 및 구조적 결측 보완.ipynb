### [1] 결측치 확인
#### (1) 결측치 수 측정

```python
df1 = origin_type_changed.copy()

n=len(df1)

missing_df=DataFrame({'missing_count':df1.isnull().sum()})
missing_df['missing_ratio']=missing_df['missing_count']/n
display(missing_df)
```

### [1] 종속변수 비율 확인

```python
import pandas as pd

dep_vars = ['dependent_ecotype', 'dependent_wage_work']

dep_summary_list = []

for col in dep_vars:
    temp = df1[col].value_counts(dropna=False).sort_index()
    ratio = df1[col].value_counts(normalize=True, dropna=False).sort_index()
    
    for val in temp.index:
        dep_summary_list.append({
            'variable': col,
            'class': val,
            'count': temp[val],
            'ratio': ratio[val],
            'percentage(%)': round(ratio[val] * 100, 2)
        })

dep_summary = pd.DataFrame(dep_summary_list)

dep_summary
```

### [2] 명목형 변수 항목별 비율

```python
categorical_cols = df1.select_dtypes(include=['category', 'object']).columns

cat_summary_list = []

for col in categorical_cols:
    counts = df1[col].value_counts(dropna=False)
    ratios = df1[col].value_counts(normalize=True, dropna=False)
    
    for val in counts.index:
        cat_summary_list.append({
            'variable': col,
            'category': val,
            'count': counts[val],
            'percentage(%)': round(ratios[val] * 100, 2)
        })

cat_summary_df = pd.DataFrame(cat_summary_list)

cat_summary_df
```

### [3] 연속형 변수

```python
from scipy.stats import skew, kurtosis
import numpy as np

numeric_cols = df1.select_dtypes(include=['int64', 'float64']).columns

num_summary_list = []

for col in numeric_cols:
    s = df1[col].dropna()
    
    if len(s) > 0:
        q1 = s.quantile(0.25)
        q3 = s.quantile(0.75)
        iqr = q3 - q1
        
        num_summary_list.append({
            'variable': col,
            'mean': round(s.mean(), 3),
            'median': round(s.median(), 3),
            'std': round(s.std(), 3),
            'IQR': round(iqr, 3),
            'skewness': round(skew(s), 3),
            'kurtosis': round(kurtosis(s), 3),
            'missing_count': df1[col].isna().sum(),
            'missing_ratio(%)': round(df1[col].isna().mean()*100, 2)
        })

num_summary_df = pd.DataFrame(num_summary_list)

num_summary_df
```
